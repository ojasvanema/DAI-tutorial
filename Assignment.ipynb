```python
plt.figure(figsize=(20,10))
plt.subplot(141)
plt.title('Gender',fontsize = 20)
df['gender'].value_counts().plot.pie(autopct="%1.1f%%")

plt.subplot(142)
plt.title('specialisation',fontsize = 20)
df['specialisation'].value_counts().plot.pie(autopct="%1.1f%%")

plt.subplot(143)
plt.title('Stream in higher secondary school',fontsize = 20)
df['hsc_s'].value_counts().plot.pie(autopct="%1.1f%%")

plt.subplot(144)
plt.title('degree stream',fontsize = 20)
df['degree_t'].value_counts().plot.pie(autopct="%1.1f%%")
plt.show()
```

* There are more males than female
* Students prefer Market and finance over marketing and HR
* Majority of the students took commerce stream in higher secondary school
* Connected to previous point as majority students took commerce , in college majority are in commerce and management but here from the data we can see that some students from other streams science or arts may have also joined commerce and management stream

```python
plt.figure(figsize=(15,5))
plt.subplot(131)
plt.title('SSC Percentage')
sns.barplot(hue='gender',x='status',y='ssc_p',data=df)

plt.subplot(132)
plt.title('HSC Percentage')
sns.barplot(hue="gender",x='status',y='hsc_p',data=df)

plt.subplot(133)
plt.title('degree Percentage')
sns.barplot(hue="gender",x='status',y='degree_p',data=df)
plt.show()
```

* SSC Percentage (First Graph):

* Both males and females who are placed have a similar SSC percentage, hovering around 70%.
* Interestingly, females who are not placed have a higher SSC percentage than males who are not placed, suggesting that SSC scores may not be a strong determinant of placement for females compared to males.

* HSC Percentage (Second Graph):

* Placed students (both male and female) have similar HSC percentages, around 70%.
* Similar to SSC, females who are not placed show a higher HSC percentage than males who are not placed. This trend continues the notion that academic performance in earlier stages (SSC and HSC) may not necessarily be linked to placement success for females.

* Degree Percentage (Third Graph):

* Among placed students, females have a slightly higher degree percentage compared to males.
* For students who are not placed, females again show a higher degree percentage than males.
This trend suggests that although females tend to have higher academic percentages, especially among non-placed individuals, it may not be translating into placements as effectively.

Fewer females are pursuing Science & Technology degrees compared to males, and only a few are in the "Others" category.

* Females show a strong preference for Commerce & Management degrees, while males dominate in Science & Technology.

```python
columns_to_plot = df[["ssc_p","ssc_b","degree_p","etest_p","mba_p","salary"]]

plt.figure(figsize=(8, 6))
for column in columns_to_plot:
    plt.figure(figsize=(6, 4))
    sns.boxplot(y=df[column])

    plt.title(f'Box Plot of {column}')
    plt.ylabel('Values')

    plt.show()
```

```python
Q1 = df["salary"].quantile(0.25)
Q3 = df["salary"].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

filtered_df = df[(df["salary"] > lower_bound) & (df["salary"] < upper_bound)]
```

```python
plt.figure(figsize=(6, 4))
sns.boxplot(y=filtered_df["salary"])

plt.title(f'Box Plot of salary')
plt.ylabel('Values')

plt.show()
```
